```python
import os
import cv2
import numpy as np
from tqdm import tqdm

import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
```

```python
model = MobileNetV2(
    weights = 'imagenet',
    include_top = False,
    pooling = "avg"
)
```

```python
def extract_features(frames_dir):
  features=[]
  frame_names = sorted(os.listdir(frames_dir))
  for name in tqdm(frame_names):
    frame_path = os.path.join(frames_dir, name)
    frame = cv2.imread(frame_path)
    if frame is None:
      continue
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame = cv2.resize(frame, (224, 224))
    frame = np.expand_dims(frame, axis=0)
    frame = preprocess_input(frame)
    feature = model.predict(frame,verbose=0)
    features.append(feature[0])
  return np.array(features)

```

```python
frames_dir = "/content/drive/MyDrive/video summarizer/data/frames"
features = extract_features(frames_dir)

print("Feature shape:", features.shape)
```

```python
np.save("/content/drive/MyDrive/video summarizer/features/features.npy", features)
```
